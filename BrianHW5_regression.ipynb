# HW5: Predict Unemployment Based on Economic Conditions
In this homework, we will test alot of what we studied about ML using the economics dataset. This dataset has 5 columns 
- psavert: personal savings rate.
- pce: personal consumption expenditures, in billions of dollars.
- uempmed: median duration of unemployment, in weeks.
- pop: total population, in thousands.
- unemploy: number of unemployed in thousands 

The goal is to build a model that predicts the unemployment based on the other features


Start by loading the data into your notebook. Form your input X (set of features) and output Y (Unemployment) to ML

```python
import pandas as pd 
import numpy as np 
df = pd.read_csv('hw5_data.csv')
df
```

```python
X = df[['psavert', 'pce', 'uempmed', 'pop']].values
Y = df['unemploy'].values
print('values for x',X)
print('values for y',Y)
```

## KMeans
Do clustering using KMeans using K=2 on X then use the model to predict the clusters for each data point

```python
from sklearn.cluster import KMeans
k= 2
kmeans = KMeans(n_clusters=k, random_state=170)
cluster_labels = kmeans.fit_predict(X)

print('Cluster labels:', cluster_labels)
```

Let's see if our clusters are meaningful. Plot a histogram of the unemployment (Y) for points that has cluster 0 and another one for points that have cluster 1. Is there some seperation? Comment on that

```python
import matplotlib.pyplot as plt
cluster_0_indices = cluster_labels ==0
cluster_1_indices = cluster_labels ==1
Y_cluster_0 = Y[cluster_0_indices]
Y_cluster_1 = Y[cluster_1_indices]

plt.figure(figsize=(10,6))
plt.hist(Y_cluster_0, bins=10, alpha=0.5, label='cluster 0')
plt.hist(Y_cluster_1, bins=10, alpha=0.5, label='cluster 1')
plt.xlabel('unemployment')
plt.ylabel('Frequency')
plt.title('Histogram of Unemployment for Clusters 0 and 1')
plt.legend()
plt.show()
```

```python
#there is a clear separation of the clusters with atleast 3 different clusters for cluster 0
```

### Split the data
Split the dataset to have 30% test data

```python
from sklearn.model_selection import train_test_split
test_size = 0.3
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=test_size, random_state=170)
print('shape of X_train:', X_train.shape)
print('shape of X_test:', X_test.shape)
print('shape of Y_train', Y_train.shape)
print('shape of Y_test', Y_test.shape)
```

### Linear Regression
Using LR, what is the square root of Mean Square Error (SMSE) for the training and the test data?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
```

```python
LinearRegression()
```

```python
Y_train_pred = linreg.predict(X_train)
Y_test_pred = linreg.predict(X_test)
```

```python
mse_train = mean_squared_error(Y_train, Y_train_pred)
mse_test = mean_squared_error(Y_test, Y_test_pred)
```

```python
smse_train = mse_train**0.5
smse_test = mse_test**0.5
```

```python
print ('SMSE for training data:', smse_train )
print('SMSE for Test Data:', smse_test)
```

## Deep Neural Networks
Using Keras, build a DNN to obtain a better SMSE than LR use at least 3 hidden layers with at least 50 neurons each. What is the MSE for training and testing data?

hint: Best performance I got from my DNN quick model was SMSE = 700 for test. Bonus if you can do better

```python
from tensorflow import keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(50,activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(50, activation= 'relu'))
model.add(Dense(50, activation= 'relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error' , metrics= ['accuracy'])

history= model.fit(X_train, Y_train, epochs=35000, batch_size=401, validation_data=(X_test, Y_test))
```

```python
Y_train_pred_dnn = model.predict(X_train)
Y_test_pred_dnn= model.predict(X_test)
```

```python
mse_train_dnn = mean_squared_error(Y_train, Y_train_pred_dnn)
mse_test_dnn = mean_squared_error(Y_test, Y_test_pred_dnn)

print('MSE for Training Data (DNN):', mse_train_dnn)
print('MSE for Test Data(DNN):', mse_test_dnn)
```

```python
smse_train_dnn = (mse_train_dnn**0.5)
smse_test_dnn = (mse_test_dnn**0.5)

print('SMSE for Test data(DNN):', smse_test_dnn)
print('SMSE for train data(DNN):', smse_train_dnn)
```

plot the loss vs epochs for train and test data

```python
plt.plot(np.arange(1, len(history.history['loss'])+1),history.history['loss'],label='Training Loss')
plt.plot(np.arange(1, len(history.history['val_loss'])+1), history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

## Random Forest
Build a random forest with `n_estimators=100`. What is the SMSE the training and the test data?

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, Y_train)
```

```python
Y_train_pred_rf = rf.predict(X_train)
Y_test_pred_rf = rf.predict(X_test)
print('pred_y_train',Y_train_pred_rf)
print('pred_y_test',Y_test_pred_rf)
```

```python
mse_train_rf = mean_squared_error(Y_train, Y_train_pred_rf)
mse_test_rf = mean_squared_error(Y_test, Y_test_pred_rf)
print('mse for train', mse_train_rf)
print('mse for test',mse_test_rf)
```

```python
smse_train_rf = (mse_train_rf**0.5)
smse_test_rf = (mse_test_rf**0.5)

print('SMSE for Training:', smse_train_rf)
print('SMSE for test:', smse_test_rf)
```
